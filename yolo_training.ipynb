```python
from glob import glob
from sklearn.model_selection import train_test_split

# image 파일들을 train, validation set으로 나눈 후에 파일이름 list화
img_list = glob('/Users/yunsu/autonomous_project/jetbot/dataset/train/images/*.jpg')

train_img_list, val_img_list = train_test_split(img_list, test_size=0.1, random_state=2000)
print(len(train_img_list), len(val_img_list))
print(len(img_list))
```

```python
# 위에서 생성한 img_list를 txt 파일로 저장
with open('/Users/yunsu/autonomous_project/jetbot/dataset/train.txt', 'w') as f: # write
  f.write('\n'.join(train_img_list) + '\n')

with open('/Users/yunsu/autonomous_project/jetbot/dataset/val.txt', 'w') as f: # write
  f.write('\n'.join(val_img_list) + '\n')
```
